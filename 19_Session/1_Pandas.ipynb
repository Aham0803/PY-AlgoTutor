```python
import pandas as pb
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago'],
    'Position': ['winner', 'first_runner_up', 'second_runner_up']
}
data
```

```python
df = pb.DataFrame(data)
df
```

```python
df[['Name','Age']] ## selecting multiple columns from the dataframe
```

```python
df
```

```python
df.loc[1] ## selecting a single row from the dataframe using loc
```

```python
df.iloc[1] ## selecting a single row from the dataframe using iloc, it will select the second row of the dataframe because iloc uses integer indexing, so it starts from 0
```

```python
df
```

Setting individual column as a index

```python
df.set_index('Position', inplace=True) ## setting the index of the dataframe to the 'Position' column, inplace=True will modify the original dataframe
```

```python
df
```

```python
df.index = ['winner','first_runner_up','second_runner_up'] ## setting the index of the dataframe to a list of values, it will replace the existing index with the new index
```

```python
df
```

```python
df.iloc[1]  ## Position          ## selecting a single row from the dataframe using iloc, it will select the second row of the dataframe because iloc uses integer indexing, so it starts from 0
```

```python
df.loc['first_runner_up'] ## labels
```

```python
del df['City']
```

```python
df
```

```python
df.drop(['Age'], axis=1) ## dropping a column from the dataframe, axis=1 means we are dropping a column, inplace=True will modify the original dataframe
```

```python
df
```

```python
new_column = {
    'Name':'NAME',
    'Age':'AGE'
}
df.rename(columns=new_column, inplace=True) ## renaming columns in the dataframe, inplace=True will modify the original dataframe

df
```

```python
df['NAME_UPPER'] = df['NAME'].str.upper() ## converting the values in the 'NAME' column to uppercase using the str.upper() method
```

```python
df 
```

Performing Data Analysis Over Some Random Data

CSV Datafile

```python
data = pb.read_csv('2_data.csv') ## reading a csv file into a dataframe using the read_csv() method
data
```

Descriptive Analysis

```python
data.info() ## Descriptive statistics of the dataframe, it will show the number of non-null values, data types and memory usage of the dataframe
```

Statistical Analysis

```python
data.describe() ## Statistical summary of the dataframe, it will show the count, mean, std, min, 25%, 50%, 75% and max values of the numerical columns in the dataframe
```

```python
data.head(2)
```

```python
data.tail(2)
```

```python
data.isnull().sum() ## checking for null values in the dataframe, it will return a dataframe of the same shape with boolean values indicating whether the value is null or not
```

```python
data.notnull().sum() ## checking for non-null values in the dataframe, it will return a dataframe of the same shape with boolean values indicating whether the value is not null or not
```

```python
data.dropna() ## dropping rows with null values from the dataframe, it will return a new dataframe with the rows containing null values removed
```

Dropping the columns which has null values

```python
data.dropna(axis=1)
```

```python
data
```

Instead of dropping just fill the Null Values with actual values
