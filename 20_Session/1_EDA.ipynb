Identifying IQR

```python
# numpy to identifying iqr
q1 = np.percentile(required_column['Radius (mean)'] , 25)
q3 = np.percentile(required_column['Radius (mean)'] , 75)
iqr = q3-q1
iqr
```

```python
#pandas to identifying iqr
q1 = required_column['Radius (mean)'].quantile(0.25)
q3 = required_column['Radius (mean)'].quantile(0.75)
iqr = q3-q1
iqr
```

```python
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
print(lower_bound)
print(upper_bound)
```

```python
final_radius = required_column[(required_column['Radius (mean)'] >= lower_bound) &(required_column['Radius (mean)'] <= upper_bound)]
final_radius.describe()
```

```python
df['final-score'] =(df['py-score'] + df['java-score'])/2
df
```

```python
df['full-name'] = df.apply(lambda x : x['name'] + ' '+ x['last-name'],axis =1)
df
```
